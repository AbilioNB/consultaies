## Processando CPC

```python
df_cpc.head()
```

```python
df_cpc.drop(['Ano da Edição','Nº de Matrículas'], axis=1,inplace=True)
```

```python
df_cpc.head()
```

```python
df_cpc.columns = ['codigo_ies','codigo_curso','cpc_continuo']
```

Conceito Enade
(Faixa)<br>

1: 0 ≤ NCc < 0,945<br>
2: 0,945 ≤ NCc < 1,945<br>
3: 1,945 ≤ NCc < 2,945<br>
4: 2,945 ≤ NCc < 3,945<br>
5: 3,945 ≤ NCc ≤ 5


```python
def calcular_faixa(nota_conitnua):
    if(nota_conitnua>=0 and nota_conitnua<0.945):
        return 1
    elif (nota_conitnua>=0.945 and nota_conitnua<1.945):
        return 2
    elif (nota_conitnua>=1.945 and nota_conitnua<2.945):
        return 3
    elif (nota_conitnua>=2.945 and nota_conitnua<3.945):
        return 4
    elif (nota_conitnua>=3.945 and nota_conitnua<=5):
        return 5
```

```python
df_cpc['cpc_faixa'] = df_cpc['cpc_continuo'].apply(calcular_faixa)
```

```python
df_cpc
```

```python
df_cpc.columns=['CODIGO_IES','CODIGO_CURSO','CPC_CONTINUO','CPC_FAIXA']
```

```python
df_cpc.to_csv('bases_processadas/CPC.csv',index=False)
```
